# Removing Outliers

```python
from scipy import stats
z = np.abs(stats.zscore(df[df.dtypes[df.dtypes != 'object'].index]))
df = df[(z<3).all(axis=1)]
```

```python
print(z)
```

```python
print(df)
```

```python
X = df.drop('quality',axis=1)
Y = df['quality']
```

```python
Y = df['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x=Y)
```

# handling imbalance data

```python
from imblearn.over_sampling import SMOTE # synthetic minority oversampling technique
smote = SMOTE()
x_smote,y_smote=smote.fit_resample(X,Y)
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=y_smote)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x_smote,y_smote,test_size=0.2,random_state=2)
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(solver='newton-cg')
```

```python
lr_model.fit(X_train,Y_train)
```

```python
lr_model.score(X_test,Y_test)*100
```

```python
y_lr_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_lr_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_lr_pred))
```

# DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(random_state=42,max_depth=3,min_samples_leaf=1,min_samples_split=2)
```

```python
dtree.fit(X_train,Y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_pred))
```

# RANDOM FOREST CLASSIFIER

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,Y_train)
```

```python
y_rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(Y_test,y_rfc_pred))
```

```python
print(confusion_matrix(Y_test,y_rfc_pred))
```

# model accuracy

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.heatmap(confusion_matrix(Y_test,y_lr_pred),annot=True)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('CONFUSION MATRIX FOR LOGISTIC REGRESSION')

plt.subplot(2,2,2)
sns.heatmap(confusion_matrix(Y_test,y_pred),annot=True)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('CONFUSION MATRIX FOR DECISION TREE')

plt.subplot(2,2,3)
sns.heatmap(confusion_matrix(Y_test,y_rfc_pred),annot=True)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('CONFUSION MATRIX FOR RANDOM FOREST CLASSIFIER')
```

```python
models = ['Logistic Regression','Decision Tree','Random Forest Classifer']
accuracy = [accuracy_score(Y_test,y_lr_pred),accuracy_score(Y_test,y_pred),accuracy_score(Y_test,y_rfc_pred)]
plt.figure(figsize=(10,8))
sns.barplot(x=models,y=accuracy)
plt.title('MODEL ACCURACY COMPARISON')
plt.xlabel('MODEL')
plt.ylabel('ACCURACY')
```

# PREDICTIVE SYSTEM

```python
input_data = (7.4,0.66,0.0,1.8,0.075,13.0,40.0,0.9978,3.51,0.56,9.4)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = rfc.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==1):
    print('GOOD QUALITY WINE')
else:
    print("BAD QUALITY WINE")
```
